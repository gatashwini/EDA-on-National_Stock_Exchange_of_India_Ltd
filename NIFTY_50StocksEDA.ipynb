# NIFTY -50 Stocks EDA

#import the required packages

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
#load the data and create dataframe

df=pd.read_csv('National_Stock_Exchange_of_India_Ltd.csv')
df.head()
```

#Description of features available in Dataset  
symbol = stocks names of nifty 50  
Open=Open price for current day  
High=Price made high for current day  
Low=Price made low for current day  
Chng=Change in rupees  
% Chng=Percentage of change from previous day close  
Volume (lacs)=Traded volume  
Turnover (crs.)=Turnover  
52w H=Stock 52 week high price  
52w L=Stock 52 week low price  
365 d % chng=Stock price percentage change from one year  
30 d % chng=Stock price percentage change from last 30days  

```python
#Columns avaible in dataset 
df.columns
```

```python
#DataTypes
df.dtypes
```

```python
#Replace the column data 
df['Open'] = df['Open'].str.replace(',','')
df['High'] = df['High'].str.replace(',','')
df['Low'] = df['Low'].str.replace(',','')
df['LTP'] = df['LTP'].str.replace(',','')
df['Turnover (crs.)'] = df['Turnover (crs.)'].str.replace(',','')
df['52w H'] = df['52w H'].str.replace(',','')
df['52w L'] = df['52w L'].str.replace(',','')
df.head()
```

```python
#change the data type of open, High,Low, LTP into float

df['Open'] = df['Open'].astype(float)
df['High'] = df['High'].astype(float)
df['Low'] = df['Low'].astype(float)
df['LTP'] = df['LTP'].astype(float)
df['Turnover (crs.)'] = df['Turnover (crs.)'].astype(float)
df['52w H'] = df['52w H'].astype(float)
df['52w L'] = df['52w L'].astype(float)
df.head()
```

```python
#Visualization of opening price with respect to Symbol
plt.figure(figsize=[15,5])
plt.xlabel='Symbol'
plt.ylabel='Open'
plt.title('Opening Price of Symbol')
plt.xticks(rotation=90)
plt.bar(df['Symbol'],df['Open'],)

plt.show()
```

```python
plt.figure(figsize=(15,7))
colors=['lightblue','salmon']
plt.xticks(rotation=90)
plt.bar(df['Symbol'],df['30 d % chng'],color=colors)
plt.title('Monthly Data of Change in Rupees')
plt.show()
```

Plotting BarPlot, we can find that majority of the companies have shown negative returns if stayed invested for a month in it.
Hence investment period should be more than a month.

```python
plt.figure(figsize=(15,7))
colors=['lightblue','salmon']
plt.xticks(rotation=90)
plt.bar(df['Symbol'],df['365 d % chng'],color=colors)
plt.title('Yearly Data of Change in Rupees')
plt.show()
```

Plotting BarPlot, we can find that majority of the 
companies have shown atleast 10% returns if stayed invested for a year in it

```python
#Visualization of Turnover with respect to Symbol
plt.figure(figsize=[15,5])
plt.xlabel='Symbol'
plt.ylabel='Turnover (crs.)'
plt.title('Turnover (crs.) of Symbol')
plt.xticks(rotation=90)
plt.bar(df['Symbol'],df['Turnover (crs.)'],)

plt.show()
```
